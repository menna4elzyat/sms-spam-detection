```python
from groq import Groq

client = Groq(api_key="gsk*****************")

def explain_with_llm(text, label, confidence):
    prompt = f"""
Explain why the following SMS is classified as {label} using simple reasoning.

SMS: {text}
Prediction: {label}
Confidence: {confidence:.2f}

Keep the explanation short and clear.
"""

    response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

    return response.choices[0].message.content
```

```python
loaded_tfidf = joblib.load("../models/tfidf_vectorizer.pkl")
loaded_tfidf_model = joblib.load("../models/balanced_svm_tfidf.pkl")
```

```python
import joblib
import os
sample_sms = "Congratulations! You won a free prize. Call now to claim."

sample_tfidf = loaded_tfidf.transform([sample_sms])

prediction = loaded_tfidf_model.predict(sample_tfidf)[0]
probability = loaded_tfidf_model.predict_proba(sample_tfidf)[0]

label = "spam" if prediction == 1 else "ham"
confidence = probability[prediction]

explanation = explain_with_llm(sample_sms, label, confidence)

print("Prediction:", label)
print("Confidence:", confidence)
print("Explanation:", explanation)
```

### LLM Explanation Result

The LLM provides a simple explanation for why the SMS was classified as spam or ham.

This improves interpretability because the system does not only return a prediction, but also explains the reason behind it.
